```python
def calculate_body_energy(time, load, decay_coefficient=0.5):
    """
    计算身体电量
    :param time: 当前时间（小时）
    :param load: 心肺负荷
    :param decay_coefficient: 衰减系数，默认为 0.5
    :return: 当前身体电量
    """
    
    initial_energy = 100
    energy = initial_energy - decay_coefficient * load * time
    # 确保电量不小于 0
    energy = max(0, energy)
    return energy

# 示例使用
current_time = 2  # 假设当前时间是 2 小时
cardiopulmonary_load = 3  # 假设心肺负荷为 3
energy = calculate_body_energy(current_time, cardiopulmonary_load)
print(f"当前身体电量: {energy}")
```

```python
import pandas as pd

# 将数据转换为 DataFrame
df = pd.read_csv('/Users/wangshuaibo/Downloads/Heart Rate 2.csv')
```

```python
df
```

```python
# 将时间戳转换为 datetime 类型
df['timestamp'] = pd.to_datetime(df['Date/Time'])
```

```python
# 按时间戳排序
df = df.sort_values(by='timestamp')
```

```python
# 重采样为 10 秒间隔
resampled_df = df.set_index('timestamp').resample('10S').mean().ffill()
```

```python
# 计算心率负荷，这里简单地将心率值乘以时间间隔（10 秒）
resampled_df['heart_load'] = resampled_df['Min (count/min)'] * 10

# 按 10 分钟分组计算心率负荷
ten_minute_load = resampled_df.groupby(pd.Grouper(freq='10T')).sum()['heart_load']
```

```python



# 计算累加心率负荷
cumulative_load = ten_minute_load.cumsum()

# 输出结果
print("每 10 分钟对应的心率负荷：")
print(ten_minute_load)
print("\n累加的心率负荷：")
print(cumulative_load)
```
