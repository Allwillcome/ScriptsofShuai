```python
import re
import jieba
from docx import Document

# 读取Word文档并返回文本内容
def read_word_document(file_path):
    doc = Document(file_path)
    text = ''
    for para in doc.paragraphs:
        text += para.text
    return text

# 分析文本并返回报告
def text_analysis(text):
    # 使用jieba进行分词，获取总字数
    words = jieba.lcut(text)
    total_words = len(words)

    # 统计逗号数
    comma_count = len(re.findall(r'，', text)) + len(re.findall(r',\s*', text))

    # 统计逗号率
    comma_rate = comma_count / total_words * 100 if total_words > 0 else 0

    # 统计问号数
    question_count = len(re.findall(r'\?\s*', text)) + len(re.findall(r'？', text))

    # 统计句子数
    sentence_count = len(re.findall(r'[。！？]\s*', text))

    # 统计段落数
    paragraph_count = len(re.findall(r'\n\n', text))

    # 统计“你”字数
    you_count = len(re.findall(r'你', text))

    # 统计“我”字数
    me_count = len(re.findall(r'我', text))

    # 统计超长逗间句数
    long_interjections_count = len(re.findall(r'.{75,}', text))

    # 统计超长句子数
    long_sentences_count = len(re.findall(r'.{192,}', text))

    # 统计超长段落数
    long_paragraphs_count = len(re.findall(r'.{251,}\n', text))

    # 统计叹号数
    exclamation_count = len(re.findall(r'！', text))

    # 输出报告
    report = f"""
# 文稿文本检测报告

指标名称 | 标记 | 指标值 | 参考区间
---------- |- | ---- | ---
逗号数      |  | {comma_count} | 139~195
逗号率      |  | {comma_rate:.1f}% | 126.2%~239.2%
问号数      |  | {question_count} | 3~14
句子数      |  | {sentence_count} | 72~106
段落数      |  | {paragraph_count} | 34~51
“你”字数    |  | {you_count} | 1~28
“我”字数    |  | {me_count} | 0~21
超长逗间句数 |  | {long_interjections_count} | 超过75个字的短句
超长句子数   |  | {long_sentences_count} | 超过192个字的句子
超长段落数   |  | {long_paragraphs_count} | 超过251个字的段落
叹号数      |  | {exclamation_count} | -
汉字总字数   |  | {total_words} | -
"""

    return report

# 主函数
def main():
    # 替换为您的Word文档路径
    word_document_path = '/Users/wangshuaibo/Downloads/200240125_13_用好 chatGPT，一个人就是一个团队.docx'
    text = read_word_document(word_document_path)
    analysis_report = text_analysis(text)
    print(analysis_report)
```

```python
main()
```
