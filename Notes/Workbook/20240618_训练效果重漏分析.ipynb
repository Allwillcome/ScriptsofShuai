```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

plt.rcParams['font.sans-serif'] = ['Heiti TC']
plt.rcParams['font.family'] = ['Heiti TC']
plt.rcParams['axes.unicode_minus'] = False

# 定义每个分类的范围
categories = {
    "恢复运动": {"x": (1.0, 2.0), "y": (0.0, 1.0)},
    "维持健康": {"x": (2.0, 3.0), "y": (1.0, 2.0)},
    "提高体能": {"x": (3.0, 3.5), "y": (0.0, 2.0)},
    "提升乳酸耐力": {"x": (3.5, 4.0), "y": (0.0, 2.0)},
    "提高心肺耐力": {"x": (4.0, 5.0), "y": (0.0, 3.0)},
    "维持速度": {"x": (0.0, 2.0), "y": (2.0, 2.5)},
    "提高速度": {"x": (2.0, 3.0), "y": (2.5, 3.0)},
    "提升无氧能力": {"x": (3.0, 5.0), "y": (3.0, 5.0)},
}

# 创建图形和子图
fig, ax = plt.subplots()

# 定义颜色
colors = [
    "red", "green", "blue", "cyan", "magenta", "yellow", "orange", "purple"
]

# 绘制每个分类的区域
for i, (category, limits) in enumerate(categories.items()):
    x_range = limits["x"]
    y_range = limits["y"]
    rect = patches.Rectangle(
        (x_range[0], y_range[0]),  # 左下角的坐标
        x_range[1] - x_range[0],  # 宽度
        y_range[1] - y_range[0],  # 高度
        linewidth=1,
        edgecolor='black',
        facecolor=colors[i],
        alpha=0.3,
        label=category
    )
    ax.add_patch(rect)

# 设置图形的范围
ax.set_xlim(0, 5)
ax.set_ylim(0, 5)

# 添加图例，放在边框外
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# 添加标签
ax.set_xlabel("有氧训练效果")
ax.set_ylabel("无氧训练效果")
ax.set_title("训练分类可视化")

# 显示图形
plt.grid(True)
plt.show()
```
