```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/Users/wangshuaibo/Downloads/20240530_智慧操场招标查询.xlsx',engine='openpyxl')
df.columns
```

```python
plt.rcParams['font.sans-serif'] = ['Heiti TC']
plt.rcParams['font.family'] = ['Heiti TC']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 统计不同跑步类型的数量
sport_type_counts = df['省份'].value_counts()

# 绘制饼状图
plt.pie(sport_type_counts, labels=sport_type_counts.index, autopct='%1.1f%%', startangle=140)

# 添加标题
plt.title('省份分布')

# 显示图形
plt.axis('equal')  # 保持圆形
plt.show()
```

```python

# 统计各省份的数量
category_counts = df['行业分类'].value_counts()

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values, color='#5fcd72')
plt.xlabel('省份')
plt.ylabel('数量')
plt.title('智慧操场招标各省份数量分布')
plt.xticks(rotation=90)
plt.show()
```

```python

# 统计各省份的数量
province_counts = df['省份'].value_counts()

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(province_counts.index, province_counts.values, color='#5fcd72')
plt.xlabel('省份')
plt.ylabel('数量')
plt.title('智慧操场招标各省份数量分布')
plt.xticks(rotation=90)
plt.show()
```

```python

# 统计各省份的数量
category_counts = df['行业分类'].value_counts()

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values, color='#5fcd72')
plt.xlabel('省份')
plt.ylabel('数量')
plt.title('智慧操场招标各省份数量分布')
plt.xticks(rotation=90)
plt.show()
```

```python
# 提取发布时间中的年份
df['Year'] = pd.to_datetime(df['发布时间']).dt.year

# 统计每年的数量
year_counts = df['Year'].value_counts().sort_index()

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(year_counts.index, year_counts.values, color='#5fcd72')
plt.xlabel('年份')
plt.ylabel('数量')
plt.title('每年发布数量分布')
plt.xticks(rotation=45)
plt.show()
```

```python
# 提取发布时间中的月份
df['Month'] = pd.to_datetime(df['发布时间']).dt.month

# 统计每月的数量
month_counts = df['Month'].value_counts().sort_index()

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(month_counts.index, month_counts.values, color='#5fcd72')
plt.xlabel('月份')
plt.ylabel('数量')
plt.title('每月发布数量分布')
plt.xticks(rotation=45)
plt.show()
```

```python

# 清除预算金额列中的单位 "万元"，并转换为数字类型
df['预算金额_万元'] = df['预算金额'].str.replace('万元', '').astype(float)

# 以 50 万为一个间隔，分组并统计数量
bins = range(0, int(df['预算金额_万元'].max()) + 50, 50)
budget_counts = pd.cut(df['预算金额_万元'], bins=bins).value_counts().sort_index()

# 创建柱状图
plt.figure(figsize=(10, 6))
plt.bar(budget_counts.index.astype(str), budget_counts.values, color='#5fcd72')
plt.xlabel('预算金额（万元）')
plt.ylabel('数量')
plt.title('预算金额分布')
plt.xticks(rotation=45)
plt.show()
```
