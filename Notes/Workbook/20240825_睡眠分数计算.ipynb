```python
import pandas as pd
import statsmodels.api as sm

# 假设你的Excel文件名为'sleep_data.xlsx'，并且它包含名为'DeepSleep', 'LightSleep', 'REMSleep', 'SleepDuration', 'SleepScore'的列
file_path = '/Users/wangshuaibo/Documents/ScriptsofShuai/Notes/ExampleData/睡眠分数计算.xlsx'

# 读取Excel文件
df = pd.read_excel(file_path, engine='openpyxl')
```

```python
# 定义自变量和因变量
X = df[['深睡时间', '浅睡时间', '快速眼动时间', '夜间睡眠时长（小时）']]
y = df['睡眠分数']

# 添加常数项，因为statsmodels的OLS模型不自动添加
X = sm.add_constant(X)

# 构建模型
model = sm.OLS(y, X)

# 拟合模型
results = model.fit()

# 打印模型的摘要
print(results.summary())

# 使用模型进行预测（例如，使用原始数据）
predictions = results.predict(X)
```
