```python
import numpy as np
from scipy.interpolate import interp1d

# 已知的跑力值和月跑量数据
known_run_powers = np.array([30, 38, 45, 54, 59, 66])
known_monthly_distances = np.array([60, 100, 200, 300, 400, 500])

# 创建一个插值函数
interpolator = interp1d(known_run_powers, known_monthly_distances, kind='linear', fill_value="extrapolate")

# 计算20到88之间的跑力值对应的月跑量
run_powers_to_interpolate = np.arange(20, 89)  # 生成20到88的跑力值数组
monthly_distances = interpolator(run_powers_to_interpolate)

# 打印结果
for power, distance in zip(run_powers_to_interpolate, monthly_distances):
    print(f"预估跑力值: {power}, 对应的月跑量: {distance:.2f}")
```

```python
import numpy as np
from scipy.optimize import curve_fit

# 已知的跑力值和月跑量数据
run_powers = np.array([20, 30, 38, 45, 54, 59, 66])
monthly_distances = np.array([40, 60, 100, 200, 300, 400, 500])

# 定义指数函数模型
def exponential_model(x, a, b, c):
    return a * np.exp(b * x) + c

# 使用curve_fit进行非线性最小二乘拟合
params, covariance = curve_fit(exponential_model, run_powers, monthly_distances)

# 输出拟合的指数函数参数
print("拟合的指数函数参数:")
print("a:", params[0])
print("b:", params[1])
print("c:", params[2])

# 创建一个模型函数来预测新的跑力值对应的月跑量
model_function = lambda x, a=params[0], b=params[1], c=params[2]: a * np.exp(b * x) + c

# 使用这个模型函数来预测新的跑力值对应的月跑量
# 例如，预测跑力值为70时的月跑量
predicted_run_power = 70
predicted_monthly_distance = model_function(predicted_run_power)

print(f"预测的跑力值为 {predicted_run_power} 时的月跑量为: {predicted_monthly_distance}")
```
